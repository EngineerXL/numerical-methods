```python
import matplotlib.pyplot as plt
import numpy as np


def draw_with_dots(a, b, xs, ys):
    fig, ax = plt.subplots(1, 1)
    correct_x = np.linspace(a, b, 100)
    correct_y = np.exp(correct_x) * correct_x**2
    ax.plot(correct_x, correct_y, "-r")
    ax.scatter(xs, ys)
    ax.plot(xs, ys, "-b")
    plt.show()


def draw(a, b, xs, ys):
    fig, ax = plt.subplots(1, 1)
    correct_x = np.linspace(a, b, 100)
    correct_y = np.exp(correct_x) * correct_x**2
    ax.plot(correct_x, correct_y, "-r")
    ax.plot(xs, ys, "-b")
    plt.show()
```
