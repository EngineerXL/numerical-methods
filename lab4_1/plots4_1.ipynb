```python
import numpy as np
import matplotlib.pyplot as plt

def draw_with_dots(l, r, xs, ys):
    fig, ax = plt.subplots(1, 1)
    correct_x = np.linspace(l, r, 100)
    correct_y = correct_x ** 2 + correct_x + correct_x * np.log(correct_x)
    ax.plot(correct_x, correct_y, "-r")
    ax.scatter(xs, ys)
    ax.plot(xs, ys, "-b")
    plt.show()

def draw(l, r, xs, ys):
    fig, ax = plt.subplots(1, 1)
    correct_x = np.linspace(l, r, 100)
    correct_y = correct_x ** 2 + correct_x + correct_x * np.log(correct_x)
    ax.plot(correct_x, correct_y, "-r")
    ax.plot(xs, ys, "-b")
    plt.show()
```
