```python
import numpy as np
import matplotlib.pyplot as plt

def draw(l, r, xs, ys):
    fig, ax = plt.subplots(1, 1)
    correct_x = np.linspace(l, r, 100)
    correct_y = np.cos(np.sin(correct_x)) + np.sin(np.cos(correct_x))
    ax.plot(correct_x, correct_y, "-r")
    ax.scatter(xs, ys)
    ax.plot(xs, ys, "-b")
    plt.show()
```

```python
l = 0.0
r = 1.0
```
